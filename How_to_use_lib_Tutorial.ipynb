```python
import MLProcessing as mlp
```

## Читаем файл ##

```python
path = 'attrs_2.csv'
df = mlp.read_file(path)
```

```python
df
```

## Поадгатавливаем данные для просмотра корреляций. Удаляем выбросы и приводим к $N_{(0, 1)}$ ##

```python
df_prepared = mlp.data_preparation(df)
```

```python
df_prepared
```

## Строим гистограммы для просмотра данных ##

```python
mlp.gistograms(df_prepared)
```

## Отрисовываем карты в координатах (X, Y, {Имя атрибута}) ##

```python
mlp.cards(df_prepared, 0.1, 1)
```

## Строим матрицу корреляций ##

```python
mlp.cor_matrix(df_prepared)
```

## Удаляем атрибуты, у которых корреляция выше 0.7 и строим повторно матрицу корреляций на новых данных, чтобы посмотреть что получилось ## 

```python
df_work = mlp.del_cors(df_prepared, 0.7)
```

```python
mlp.cor_matrix(df_work)
```

## Подготавливаем данные к кластеризации, удаляем параметр X, Y ##

```python
df_for_clust = mlp.prepare_classters(df_work)
```

## Сжимаем данные в $\mathbb{R}^2$ ##

```python
pca_2 = mlp.pca(df_for_clust, 2)
```

## Создаем класстеры на данных pca и со всеми атрибутами ##

```python
cl_2 = mlp.clasters(df_work, 3)
cl_pca_2 = mlp.clasters(pca_2, 3)
```

## Отрисовываем класстеры ##

```python
mlp.pca_classter_labels(df_work, pca_2, cl_pca_2)
mlp.classter_labels(df_work, cl_2)
```

## Отрисовываем облоко, полученное при помощи pca ##

```python
mlp.chip(pca_2)
```
